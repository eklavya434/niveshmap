# Notebook 01: Price Proxy Feasibility Exploration

## Analytical Question
Can comparable historical price proxies be constructed across mature and emerging NCR localities?

## Methodology
We load the locality quarterly panel dataset `data/processed/phase0_quarterly_panel.csv` and compare price coverage and trends.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load panel dataset
panel_df = pd.read_csv("../data/processed/phase0_quarterly_panel.csv")

# Check missingness per locality
missingness = panel_df.groupby("locality_id")["price_proxy"].apply(lambda x: x.isna().mean() * 100)
print("Percentage of missing quarters in price_proxy:")
print(missingness)
```

### Plotting Price Trends for Mature vs Emerging Localities

```python
localities_to_plot = ["DEL_DWR_SEC10", "NOI_SEC150", "YEX_SEC22D"]
plt.figure(figsize=(10, 5))

for loc_id in localities_to_plot:
    loc_df = panel_df[panel_df["locality_id"] == loc_id]
    # Drop NA to show gaps in the trend line
    plt.plot(loc_df["quarter"], loc_df["price_proxy"], marker='o', label=loc_id)

plt.title("Historical Price Proxy Comparison (Delhi vs Noida vs YEIDA)")
plt.xlabel("Quarter")
plt.ylabel("Composite Price Proxy (INR/sqft)")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.savefig("../reports/figures/price_proxy_trends.png")
plt.show()
```

## Findings & Limitations
- **Findings**: Mature areas like Dwarka have a continuous trend based on baselines. Emerging areas like YEIDA have sparse signals (45% missingness) because RERA registrations only started recently, making direct forecasting difficult.
- **Limitations**: Circle rates represent step-functions (infrequent adjustments) and do not capture real-time quarterly fluctuations.

## Decision
We must normalize localized prices using quarterly baselines (NHB Residex) to construct smooth, comparable trends for forecast-eligible areas.
