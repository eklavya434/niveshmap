# Notebook 00: Data Source Audit Exploration

## Analytical Question
What is the feasibility and automation status of public data sources across Delhi NCR jurisdictions?

## Methodology
We load the compiled source audit data from `data/processed/source_audit.csv` and analyze distributions of accessibility, formats, and automation feasibility.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Ensure output directories exist
os.makedirs("../reports/figures", exist_ok=True)

# Load audit results
audit_df = pd.read_csv("../data/processed/source_audit.csv")
audit_df[["source_id", "source_name", "geography", "access_method", "status", "automation_feasibility"]].head()
```

### 1. Automation Feasibility Distribution

```python
feasibility_counts = audit_df["automation_feasibility"].value_counts()
plt.figure(figsize=(6, 4))
feasibility_counts.plot(kind="bar", color=["tomato", "skyblue", "gold"])
plt.title("Source Automation Feasibility in Delhi NCR")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig("../reports/figures/automation_feasibility.png")
plt.show()
```

### 2. Status Breakdown

```python
status_counts = audit_df["status"].value_counts()
print("Source status distributions:")
print(status_counts)
```

## Findings & Limitations
- **Findings**: The majority of RERA portals (`up_rera`, `hrera_gurugram`) protect detailed project pages with CAPTCHAs, preventing direct web scraping. They are marked as `manual_download_required`.
- **Limitations**: PDF circle rate schedules require custom text parsing patterns per district (Delhi vs Noida vs Gurugram).
