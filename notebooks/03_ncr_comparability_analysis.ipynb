# Notebook 03: Cross-NCR Comparability Analysis

## Analytical Question
How comparable are real estate price and transit features across the three state/UT jurisdictions (Delhi, Uttar Pradesh, Haryana)?

## Methodology
We load `data/processed/locality_feasibility.csv` and compare the readiness distribution across regions.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load locality feasibility results
df = pd.read_csv("../data/processed/locality_feasibility.csv")

# Pivot or group to see score distribution by region
regional_stats = df.groupby("region")["data_readiness_score"].agg(["count", "mean", "min", "max"])
print("Data Readiness Score statistics by region:")
print(regional_stats)
```

### Visualizing Readiness Across Regions

```python
plt.figure(figsize=(10, 5))
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f']

plt.bar(df["locality_id"], df["data_readiness_score"], color=colors[:len(df)])
plt.axhline(75.0, color="green", linestyle="--", label="Full Forecast Threshold (75)")
plt.axhline(50.0, color="orange", linestyle="--", label="Limited Forecast Threshold (50)")
plt.title("Locality Data Readiness Scores Across Delhi NCR")
plt.ylabel("Readiness Score")
plt.xlabel("Locality ID")
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.savefig("../reports/figures/locality_readiness_comparison.png")
plt.show()
```

### Classification Eligibility Matrix

```python
eligibility_matrix = pd.crosstab(df["region"], df["readiness_class"])
print("Localities eligibility matrix:")
print(eligibility_matrix)
```

## Findings & Limitations
- **Findings**: UP and Haryana regions have a large proportion of full/limited forecasting eligibility because RERA filing disclosures and large master-planned sectors (e.g. Sector 150, Sector 82 Gurugram) provide richer transaction-level signals. Delhi localities generally suffer from lack of RERA-style project filings and static circle rates.
- **Limitations**: Differences in valuation practices (Uttar Pradesh circle rates vs. Haryana collector rates) necessitate indexing/normalizing values rather than modeling raw absolute values directly.

## Recommendation
Construct the model using indexed trend growths normalized to base year (2019) rather than raw absolute price values to ensure comparability across different administrative jurisdictions.
